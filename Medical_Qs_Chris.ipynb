```
import pandas as pd 
import io 
  
df = pd.read_csv(io.BytesIO(uploaded['CUS635medicalquestions.csv']),encoding='latin-1') 
df.head()
```

```
#Removing Excess Data
df = df.drop(['question_id', 'question', 'answer_id', 'answer', 'answer_type', 'wrong_answer', 'wrong_answer_type', 'url', 'source'], axis=1)
```

```
df.head()
```

```
#Number of Medical Questions total
index = df.index
number_of_rows = len(index)
print(number_of_rows)
```

```
#Remove questions without a '?'
df = df[~-df['title'].str.contains('\?')]
```

```
#Remove questions with a '('
df = df[-df['title'].str.contains('\(')]
```

```
#Remove questions with a '"'
df = df[-df['title'].str.contains('\"')]
```

```
#Removing excess
df = df.drop(df.index[114:415])
```

```
#Final Number of questions
index = df.index
number_of_rows = len(index)
print(number_of_rows)
```
