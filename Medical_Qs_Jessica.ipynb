## Pregnancy Questions
Source: https://www.parentclub.scot/articles/common-pregnancy-questions

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.parentclub.scot/articles/common-pregnancy-questions'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
rows = soup.select('h3')
```

```python
rows = list(rows)
questions = []
for row in rows:
    row = str(row)
    parse = row.split('>')
    a = parse[1].split('<', 1)
    questions.append(a[0])
```

```python
questions = questions[0:-1]
```

```python
questions
```

```python
import pandas as pd

data = pd.DataFrame()

data['questions'] = questions
```

```python
data
```

```python
data.drop(8, inplace = True)
```

```python
data
```

## Hospice Care Questions
Source: https://www.nia.nih.gov/health/frequently-asked-questions-about-hospice-care

```python
URL = 'https://www.nia.nih.gov/health/frequently-asked-questions-about-hospice-care'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
rows = list(soup.select('h2'))
questions2 = []
for row in rows:
    a = str(row)
    if '?' in a:
        parse = a.split('>')
        parse = parse[1].split('<')
        questions2.append(parse[0])
questions2
```

```python
new_qs = list(data['questions']) + questions2

data = pd.DataFrame()
data['questions'] = new_qs
```

```python
data
```

## Common Health Questions
Source: https://www.nhs.uk/common-health-questions/

```python
URL = 'https://www.nhs.uk/common-health-questions/'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
links = []
for link in soup.find_all('a'):
    if 'common-health' in str(link.get('href')):
        links.append('https://www.nhs.uk/' + str(link.get('href')))

links
    
```

```python
questions3 = []
for link in links:
    URL = link
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    text = soup.get_text()
    text = text.split('\n')
    for item in text:
        if '?' in item:
            questions3.append(item)
    
```

```python
questions3
```

```python
for question in questions3:
    if 'NHS' in question:
        questions3.remove(question)
```

```python
new_qs = list(data['questions']) + questions3

data = pd.DataFrame()
data['questions'] = new_qs
```

```python
data
```

## Headache FAQs
Source: https://stanfordhealthcare.org/medical-conditions/brain-and-nerves/headache/faqs.html

```python
URL = 'https://stanfordhealthcare.org/medical-conditions/brain-and-nerves/headache/faqs.html'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
questions4 = []
text = soup.get_text()
text = text.split('\n')
for item in text:
    if '?' in item:
        questions4.append(item.lstrip())
```

```python
questions4 = questions4[6:]
for question in questions4:
    if len(question) > 200:
        questions4.remove(question)
questions4
```

```python
new_qs = list(data['questions']) + questions4

data = pd.DataFrame()
data['questions'] = new_qs
data
```
