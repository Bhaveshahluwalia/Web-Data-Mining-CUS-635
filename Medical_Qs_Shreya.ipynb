```python
import requests
from bs4 import BeautifulSoup

URL = 'https://my.clevelandclinic.org/health/diseases/14633-abnormal-menstruation-periods'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
rows = soup.select('h3')
rows
```

```python
rows = list(soup.select('h3'))
ques1 = []
for row in rows:
    i = str(row)
    if '?' in i:
        parse = i.split('>')
        parse = parse[1].split('<')
        ques1.append(parse[0])

ques1 = ques1[0:-1]
ques1
```

```python
import pandas as pd

df = pd.DataFrame()

df['Questions'] = ques1
df
```
