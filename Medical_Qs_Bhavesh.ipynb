```python
! pip install beautifulsoup4
import requests
from bs4 import BeautifulSoup
import io 
import pandas as pd
```

```python
URL = "https://study.com/learn/disorders-and-treatment-in-health-questions-and-answers.html"
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
rows = soup.select("div > a> div> div> div")
```

```python
len(rows)
```

```python
rows
```

```python
allquestions1 = []
for row in rows:
    a = str(row)
    if '?' in a:
        parse = a.split('>')
        parse = parse[1].split('<')
        parse = parse[0].split(' A)')
        parse = parse[0].split(' A.')
        parse = parse[0].split('(a)')
        parse = parse[0].split('a.')
        parse = parse[0].split('a)')
        parse = parse[0].split('a-')
        parse = parse[0].split('TRUE')
        parse = parse[0].split('true')
        parse = parse[0].split('True')
        parse = parse[0].split('False')
        parse = parse[0].split('FALSE')
        parse = parse[0].split('false')
        parse = parse[0].split('Regarding')

        allquestions1.append(parse[0])
allquestions1
```

```python
len(allquestions1)
```

```python
#making dataframe 
import pandas as pd
data1 = pd.DataFrame()

data1['questions'] = allquestions1
data1
```

## Dataset 2 from follwing link https://www.insider.com/common-pregnancy-questions-answers-2018-9

```python
URL = 'https://www.insider.com/common-pregnancy-questions-answers-2018-9'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
rows2 = list(soup.select('h2')) #selecting the particular tag parameter in the html file <>
```

```python
rows2
```

```python
allquestions2 = []
for row in rows2:
    a = str(row)
    if '?' in a:
        parse = a.split('>')
        parse = a.split('. ')
        parse = parse[1].split('<')
        allquestions2.append(parse[0])
allquestions2
```

```python
add_more_question = list(data1['questions']) + allquestions2
data1 = pd.DataFrame()
data1['questions'] = add_more_question
data1['questions']= data1['questions'].str.strip(' ') #removing the starting numbers in each row 
data1
```
